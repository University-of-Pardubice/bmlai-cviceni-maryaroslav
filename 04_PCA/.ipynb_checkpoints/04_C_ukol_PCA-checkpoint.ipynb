# Příprava dat
- vyberte X promenné bathrooms', 'bedrooms', 'livingArea','age','price'
- proměnné standardizujte

```python
pos = 1
fig = plt.figure(figsize=(8, 12))
for i in data.columns:
    ax = fig.add_subplot(7,2, pos)
    pos = pos + 1
    sns.histplot(data[i],ax=ax, kde=True)
```

```python
def rescale(X):
    mean = X.mean()
    std = X.std()
    scaled_X = [(i - mean)/std for i in X]
    return pd.Series(scaled_X)
```

```python
data_std = pd.DataFrame(columns=data.columns)
for i in data.columns:
    data_std[i] = rescale(data[i])
```

```python
data_std.describe()
```

```python
pos = 1
fig = plt.figure(figsize=(8,12))
for i in data_std.columns:
    ax = fig.add_subplot(7,2, pos)
    pos = pos + 1
    sns.histplot(data_std[i],ax=ax, kde=True)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)
sns.heatmap(data_std.corr(),annot=True)
```

# PCA
- Proveďte PCA
- zobrazte korelační matici

```python
import numpy as np
from numpy.linalg import inv
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
X = data.select_dtypes(include=['float64', 'int64'])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
corr_matrix = pd.DataFrame(X_pca).corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm')
plt.title('Korelační matice PCA komponent')
plt.show()
```

# Lineární model
- Vytvořte a natrénujte Lineární model pro PCA proměnné
- Zobrazte R2 a RMSE

```python
explained_variance = pca.explained_variance_ratio_
cumulative_variance = explained_variance.cumsum()
n_components = (cumulative_variance < 0.95).sum() + 1
```

```python
X_reduced = X_pca[:, :n_components]
```

```python
y = data.iloc[:, -1]
```

```python
model = LinearRegression()
model.fit(X_reduced, y)
y_pred = model.predict(X_reduced)
```

```python
from math import sqrt
rmse = sqrt(mean_squared_error(y, y_pred))
```

```python
print(f"R²: {r2:.3f}")
print(f"RMSE: {rmse:.3f}")
```
